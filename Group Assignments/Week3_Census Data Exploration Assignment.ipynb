### Plot Data
Now, I have the count of disasters by disaster types.
In this section, I am taking following steps to create a graph:
1. Put the number of disasters happend in 1970,　1980, 1990, 2000, 2010, and 2020 for each disaster type by using "y_labels_"
2. Create a plot for each disaster type with "plt.plot"
3. Add the year on x axis as a measurement
4. Recreate a plot by adding information about x axis, title, and legend

#### Put the number of disasters happend in 1970,　1980, 1990, 2000, 2010, and 2020 for each disaster type by using "y_labels_"

```python
y_labels_flood = [1,1,2,3,7,25 ]
y_labels_earthquake =[0,1,3,5,2,1]
y_labels_landslide = [0,1,0,4,1,1]
y_labels_volcanic = [0,1,1,0,2,2]
y_labels_epidemic = [0,0,1,2,0,0]
y_labels_storm = [0,1,0,0,0,0]
y_labels_wildfire = [0,0,0,1,0,0]
```

#### Create a plot for each disaster type with "plt.plot"

```python
plt.plot(y_labels_flood, label='flood')
plt.plot(y_labels_earthquake, label='earthquake')
plt.plot(y_labels_landslide, label='landslidered')
plt.plot(y_labels_volcanic, label='volcanic')
plt.plot(y_labels_epidemic, label='epidemicbrown')
plt.plot(y_labels_storm, label='stormpurple')
plt.plot(y_labels_wildfire, label='wildfire')

plt.show()
```

#### Add the year on x axis as a measurement

```python
x_labels = ['1970','1980','1990','2000','2010','2020']
```

#### Recreate a plot by adding information about x axis, title, and legend

```python
plt.plot(x_labels, y_labels_flood, label='flood')
plt.plot(x_labels, y_labels_earthquake, label='earthquake')
plt.plot(x_labels, y_labels_landslide, label='landslidered')
plt.plot(x_labels, y_labels_volcanic, label='volcanic')
plt.plot(x_labels, y_labels_epidemic, label='epidemicbrown')
plt.plot(x_labels, y_labels_storm, label='stormpurple')
plt.plot(x_labels, y_labels_wildfire, label='wildfire')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.title('Disaster Trend in the Last 50 Years')

plt.show()
```

**About the Graph**

First, the freaquencey of **flood** has been drastically increasing, especially this past 10 years. One of the reasons for this increase would be climate change. Rising temperature make more water vapor which contributes to more cloud formation. Therefore, climate change can trigger flood with extreme precipitation.

Second, there was an increase in **earthquake** around 2000. In fact, there was Indian Ocean earthquake in 2004. Therefore, this increase in earthquake from 1980 to 2000 could be considered to be a foreshock of the earthquake in 2004. Along with the trend of earthquake, trend of **landlide** is very slimiar to that of earthquake. Thus, those landlide might be mainly triggered by the shake casued by an earthquake.
